```python
import itertools
from sklearn.metrics import confusion_matrix

# Our function needs a different name to sklearn's plot_confusion_matrix
def make_confusion_matrix(y_true, y_pred, classes=None, figsize=(10, 10), text_size=15):
  """Makes a labelled confusion matrix comparing predictions and ground truth labels.

  If classes is passed, confusion matrix will be labelled, if not, integer class values
  will be used.

  Args:
    y_true: Array of truth labels (must be same shape as y_pred).
    y_pred: Array of predicted labels (must be same shape as y_true).
    classes: Array of class labels (e.g. string form). If `None`, integer labels are used.
    figsize: Size of output figure (default=(10, 10)).
    text_size: Size of output figure text (default=15).

  Returns:
    A labelled confusion matrix plot comparing y_true and y_pred.

  Example usage:
    make_confusion_matrix(y_true=test_labels, # ground truth test labels
                          y_pred=y_preds, # predicted labels
                          classes=class_names, # array of class label names
                          figsize=(15, 15),
                          text_size=10)
  """
  # Create the confustion matrix
  cm = confusion_matrix(y_true, y_pred)
  cm_norm = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis] # normalize it
  n_classes = cm.shape[0] # find the number of classes we're dealing with

  # Plot the figure and make it pretty
  fig, ax = plt.subplots(figsize=figsize)
  cax = ax.matshow(cm, cmap=plt.cm.Blues) # colors will represent how 'correct' a class is, darker == better
  fig.colorbar(cax)

  # Are there a list of classes?
  if classes:
    labels = classes
  else:
    labels = np.arange(cm.shape[0])

  # Label the axes
  ax.set(title="Confusion Matrix",
         xlabel="Predicted label",
         ylabel="True label",
         xticks=np.arange(n_classes), # create enough axis slots for each class
         yticks=np.arange(n_classes),
         xticklabels=labels, # axes will labeled with class names (if they exist) or ints
         yticklabels=labels)

  # Make x-axis labels appear on bottom
  ax.xaxis.set_label_position("bottom")
  ax.xaxis.tick_bottom()

  # Set the threshold for different colors
  threshold = (cm.max() + cm.min()) / 2.

  # Plot the text on each cell
  for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, f"{cm[i, j]} ({cm_norm[i, j]*100:.1f}%)",
             horizontalalignment="center",
             color="white" if cm[i, j] > threshold else "black",
             size=text_size)
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def compute_performance(y_true, y_pred, return_all=False):
    """
    prints different performance matrics like  Accuracy, Recall (macro), Precision (macro), and F1 (macro).
    This also display Confusion Matrix with proper X & Y axis labels.
    Also, returns F1 score

    Args:
        y_true: numpy array or list
        y_pred: numpy array or list


    Returns:
        float
    """

    # Replace "pass" statement with your code
    model_accuracy = accuracy_score(y_true, y_pred) * 100
    # Calculate model precision, recall and f1 score using "weighted" average
    model_precision, model_recall, model_f1, _ = precision_recall_fscore_support(y_true, y_pred, average="macro")
    model_results = {"accuracy": model_accuracy,
                     "precision": model_precision,
                     "recall": model_recall,
                     "f1": model_f1}
    print(model_results)
    make_confusion_matrix(y_true=y_true,
                      y_pred=y_pred,
                      classes=[0, 1],
                      figsize=(10, 10),
                      text_size=10)
    if return_all is True:
      return model_results
    else:
      return model_results['f1']
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

def get_lemma(text):
  result = ''
  for token in nlp(text):
    if nlp.vocab[token.lemma_].is_stop == False:
      result = result + token.lemma_ + ' '
  return result
```

```python
from sklearn.model_selection import train_test_split

def get_train_valid_data(train_file, val_file, train_size=1, lemmatize=True):
    train_df = pd.read_csv(train_file)

    train_df = train_df.sample(frac=1, random_state=seed)
    valid_df = pd.read_csv(val_file)

    train_df['label'].replace(to_replace=['NOT', 'OFF'], value=[0, 1], inplace=True)
    valid_df['label'].replace(to_replace=['NOT', 'OFF'], value=[0, 1], inplace=True)


    train_df = train_df.sample(frac=1, random_state=seed)

    if train_size > 0 and train_size < 1:
        train_sentences, test_sentences, train_labels, test_labels = train_test_split(train_df["tweet"].to_numpy(),
                                                                            train_df["label"].to_numpy(),
                                                                            test_size=1-train_size,
                                                                            random_state=seed,
                                                                            stratify=train_df['label'])
    else:
        train_sentences = train_df["tweet"].to_numpy()
        train_labels = train_df["label"].to_numpy()

    # Lemmatize train sentences
    if lemmatize == True:
      print('Lemmatizing train...')
      for index in range(0, len(train_sentences)):
        train_sentences[index] = get_lemma(train_sentences[index])
    #print(train_sentences[0:5])


    val_sentences = valid_df["tweet"].to_numpy()
    val_labels = valid_df["label"].to_numpy()
    # Lemmatize validation sentences
    if lemmatize == True:
      print('Lemmatizing val...')
      for index in range(0, len(val_sentences)):
        val_sentences[index] = get_lemma(val_sentences[index])
    #print(val_sentences)
    print('Training samples:', len(train_sentences))
    print('Validation samples:', len(val_sentences))
    unique, counts = np.unique(train_labels, return_counts=True)
    print(unique[0], counts[0], (counts[0]/len(train_labels))*100,'%')
    print(unique[1], counts[1], (counts[1]/len(train_labels))*100,'%')
    return train_sentences, val_sentences, train_labels, val_labels
```

```python
def get_test_data(test_file, lemmatize=True):
    test_df = pd.read_csv(test_file)
    test_df['label'].replace(to_replace=['NOT', 'OFF'], value=[0, 1], inplace=True)

    test_ids = test_df["id"].to_numpy()
    test_sentences = test_df["tweet"].to_numpy()
    test_labels = test_df["label"].to_numpy()
    if lemmatize == True:
      for index in range(0, len(test_sentences)):
        test_sentences[index] = get_lemma(test_sentences[index])

    return test_sentences, test_labels, test_ids
```

```python
import datetime
def create_tensorboard_callback(dir_name, experiment_name):
  """
  Creates a TensorBoard callback instand to store log files.
  Stores log files with the filepath:
    "dir_name/experiment_name/current_datetime/"
  Args:
    dir_name: target directory to store TensorBoard log files
    experiment_name: name of experiment directory (e.g. efficientnet_model_1)
  """
  log_dir = dir_name + "/" + experiment_name + "/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
  tensorboard_callback = tf.keras.callbacks.TensorBoard(
      log_dir=log_dir
  )
  print(f"Saving TensorBoard log files to: {log_dir}")
  return tensorboard_callback
```

```python
# Plot the validation and training data separately
import matplotlib.pyplot as plt
def plot_loss_curves(history):
  """
  Returns separate loss curves for training and validation metrics.
  """
  loss = history.history['loss']
  val_loss = history.history['val_loss']

  accuracy = history.history['accuracy']
  val_accuracy = history.history['val_accuracy']

  epochs = range(len(history.history['loss']))

  # Plot loss
  plt.plot(epochs, loss, label='training_loss')
  plt.plot(epochs, val_loss, label='val_loss')
  plt.title('Loss')
  plt.xlabel('Epochs')
  plt.legend()

  # Plot accuracy
  plt.figure()
  plt.plot(epochs, accuracy, label='training_accuracy')
  plt.plot(epochs, val_accuracy, label='val_accuracy')
  plt.title('Accuracy')
  plt.xlabel('Epochs')
  plt.legend();
```

## Training Method 1 | Feed-forward neural network | Baseline


```python
def train_method1(train_file, val_file, model_dir, train_size=1):
    """
     Takes train_file, val_file and model_dir as input.
     It trained on the train_file datapoints, and validate on the val_file datapoints.
     While training and validating, it print different evaluataion metrics and losses, wheverever necessary.
     After finishing the training, it saved the best model in the model_dir.

     ADD Other arguments, if needed.

    Args:
        train_file: Train file name
        val_file: Validation file name
        model_dir: Model output Directory
    """
    tf.random.set_seed(seed)
    train_sentences, val_sentences, train_labels, val_labels = get_train_valid_data(train_file, val_file, train_size)
    #print(train_sentences, train_labels)

    #max_length = round(sum([len(i.split()) for i in train_sentences])/len(train_sentences))
    max_length = 64
    max_vocab_length = 10000 # max number of words to have in the vocabulary

    text_vectorizer = TextVectorization(max_tokens=None,
                                    output_mode="int",
                                    output_sequence_length=max_length,
                                    standardize="lower_and_strip_punctuation",
                                    split="whitespace")
    text_vectorizer.adapt(train_sentences)
    pickle.dump({'config': text_vectorizer.get_config(),
             'weights': text_vectorizer.get_weights()}
            , open(os.path.join(model_dir, "vectorizer.pkl"), "wb"))

    embedding = layers.Embedding(input_dim=text_vectorizer.vocabulary_size(), # set input shape
                             output_dim=128, # set size of embedding vector
                             embeddings_initializer="uniform", # default, intialize randomly
                             input_length=max_length, # how long is each input
                             name="embedding_1")


    inputs = layers.Input(shape=(1,), dtype="string") # inputs are 1-dimensional strings
    x = text_vectorizer(inputs) # turn the input text into numbers
    x = embedding(x) # create an embedding of the numerized numbers
    #x = layers.Embedding(input_dim=vocab_size, output_dim=emdedding_size, weights=[pretrained_weights])(x)
    #x = layers.GlobalAveragePooling1D()(x) # lower the dimensionality of the embedding
    x = layers.GlobalMaxPool1D()(x)
    outputs = layers.Dense(1, activation="sigmoid")(x) # create the output layer, want binary outputs so use sigmoid activation
    model_1 = tf.keras.Model(inputs, outputs, name="model_1_dense") # construct the model

    # Compile model
    model_1.compile(loss="binary_crossentropy",
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

    print(model_1.summary())

    # Fit the model
    model_1_history = model_1.fit(train_sentences, # input sentences can be a list of strings due to text preprocessing layer built-in model
                              train_labels,
                              epochs=5,
                              validation_data=(val_sentences, val_labels),
                              )
    # callbacks=[create_tensorboard_callback(dir_name='trial_models', experiment_name="simple_dense_model_"+str(split_size*100))]
    # Check the results
    print(model_1.evaluate(val_sentences, val_labels))
    model_1.save(os.path.join(model_dir, 'model'))
    return model_1_history
```

## Testing Method 1

```python
def test_method1(test_file, model_file, output_dir):
    """
     take test_file, model_file and output_dir as input.
     It loads model and test of the examples in the test_file.
     It prints different evaluation metrics, and saves the output in output directory

     ADD Other arguments, if needed

    Args:
        test_file: test file name
        model_file: model file name
        output_dir: Output Directory
    """
    test_sentences, test_labels, test_ids = get_test_data(test_file)
    # Make predictions
    #model_file = loaded_model_6_SavedModel = tf.keras.models.load_model("MODEL_1_25_DIRECTORY/model")
    y_pred_probs = model_file.predict(test_sentences)
    y_pred = tf.squeeze(tf.round(y_pred_probs))
    results = compute_performance(y_true=test_labels, y_pred=y_pred, return_all=True)
    output = pd.DataFrame(columns=['id', 'tweet', 'label', 'out_label'])
    output['id'] = test_ids
    output['tweet'] = test_sentences
    output['label'] = test_labels
    output['out_label'] = y_pred
    output['label'].replace(to_replace=[0, 1], value=['NOT', 'OFF'], inplace=True)
    output['out_label'].replace(to_replace=[0, 1], value=['NOT', 'OFF'], inplace=True)
    output.to_csv(output_dir)
    return results
```

```python
def train_method2(train_file, val_file, model_dir, train_size=1):
    """
     Takes train_file, val_file and model_dir as input.
     It trained on the train_file datapoints, and validate on the val_file datapoints.
     While training and validating, it print different evaluataion metrics and losses, wheverever necessary.
     After finishing the training, it saved the best model in the model_dir.

     ADD Other arguments, if needed.

    Args:
        train_file: Train file name
        val_file: Validation file name
        model_dir: Model output Directory
    """
    tf.random.set_seed(seed)
    train_sentences, val_sentences, train_labels, val_labels = get_train_valid_data(train_file, val_file, train_size)
    #print(train_sentences, train_labels)

    #max_length = round(sum([len(i.split()) for i in train_sentences])/len(train_sentences))
    max_length = 64
    max_vocab_length = 10000 # max number of words to have in the vocabulary. Not using

    text_vectorizer = TextVectorization(max_tokens=None, # how many words in the vocabulary
                                    standardize="lower_and_strip_punctuation", # how to process text
                                    split="whitespace", # how to split tokens
                                    ngrams=None, # create groups of n-words?
                                    output_mode="int", # how to map tokens to numbers
                                    output_sequence_length=None) # how long should the output sequence of tokens be?
                                    # pad_to_max_tokens=True) # Not valid if using max_tokens=None
    text_vectorizer.adapt(train_sentences)
    pickle.dump({'config': text_vectorizer.get_config(),
             'weights': text_vectorizer.get_weights()}
            , open(os.path.join(model_dir, "vectorizer.pkl"), "wb"))


    # Finds custom embeddings while training the model via back-propagation
    embedding = layers.Embedding(input_dim=text_vectorizer.vocabulary_size(),
                                     output_dim=128,
                                     embeddings_initializer="uniform",
                                     input_length=max_length,
                                     name="embedding_2")


    # Create CNN model
    inputs = layers.Input(shape=(1,), dtype="string")
    x = text_vectorizer(inputs)
    x = embedding(x)
    x = layers.Conv1D(filters=32, kernel_size=5, activation="relu")(x)
    x = layers.GlobalMaxPool1D()(x)
    x = layers.Dense(64, activation="relu")(x) # optional dense layer
    outputs = layers.Dense(1, activation="sigmoid")(x)
    model_2 = tf.keras.Model(inputs, outputs, name="model_5_Conv1D")

    # Compile Conv1D model
    model_2.compile(loss="binary_crossentropy",
                    optimizer=tf.keras.optimizers.Adam(),
                    metrics=["accuracy"])
    model_2.summary()

    # Fit model
    model_2_history = model_2.fit(train_sentences,
                              train_labels,
                              epochs=5,
                              validation_data=(val_sentences, val_labels))

    # Check the results
    print(model_2.evaluate(val_sentences, val_labels))
    model_2.save(os.path.join(model_dir, 'model'))
    #with open(os.path.join(model_dir, 'model.sav'), 'wb') as fin:
      #pickle.dump(model_2, fin)
    return model_2_history
```

```python
import tensorflow_hub as hub
def train_method3(train_file, val_file, model_dir, train_size=1):
    """
     Takes train_file, val_file and model_dir as input.
     It trained on the train_file datapoints, and validate on the val_file datapoints.
     While training and validating, it print different evaluataion metrics and losses, wheverever necessary.
     After finishing the training, it saved the best model in the model_dir.

     ADD Other arguments, if needed.

    Args:
        train_file: Train file name
        val_file: Validation file name
        model_dir: Model output Directory
    """
    tf.random.set_seed(seed)
    train_sentences, val_sentences, train_labels, val_labels = get_train_valid_data(train_file, val_file, train_size, lemmatize=False)

    #preprocessor = hub.KerasLayer("https://tfhub.dev/google/universal-sentence-encoder-cmlm/multilingual-preprocess/2")
    #https://tfhub.dev/google/nnlm-en-dim128/2
    #https://tfhub.dev/google/universal-sentence-encoder/4
    #https://tfhub.dev/google/universal-sentence-encoder-large/5
    embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder-large/5") # load Universal Sentence Encoder

    #preprocessed_text = preprocessor(train_sentences)
    #input_text = embed(preprocessed_text)['pooled_output']


    callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)

    sentence_encoder_layer = hub.KerasLayer(embed,
                                            #output_shape=128,
                                            input_shape=[], # shape of inputs coming to the model
                                            dtype=tf.string, # data type of inputs coming to the USE layer
                                            trainable=False, # keep the pretrained weights
                                            name="USE")
    model_3 = tf.keras.Sequential([
      sentence_encoder_layer, # take in sentences and then encode them into an embedding
      layers.Dense(64, activation="relu"),
      layers.Dense(1, activation="sigmoid")
    ], name="model_3_USE")

    # Compile model
    model_3.compile(loss="binary_crossentropy",
                    optimizer=tf.keras.optimizers.Adam(),
                    metrics=["accuracy"])

    model_3.summary()

    # Fit model
    model_3_history = model_3.fit(train_sentences,
                              train_labels,
                              epochs=5,
                              validation_data=(val_sentences, val_labels),
                              callbacks=[callback])

    # callbacks=[create_tensorboard_callback(dir_name='trial_models', experiment_name="simple_dense_model_"+str(split_size*100))]
    # Check the results
    print(model_3.evaluate(val_sentences, val_labels))
    model_3.save(os.path.join(model_dir, 'model'))
    return model_3_history
```
